```python
from tkinter import *

def click(event):
    global dispval
    text = event.widget.cget("text")
    if text == "=":
        if dispval.get().isdigit():
            value = int(dispval.get())
        else:
            try:
                value=eval(inp.get())
            except Exception as e:
                print(e)
                value = "ERROR"
        dispval.set(value)
        inp.update()
    elif text == "C":
        dispval.set("")
        inp.update()
    else:
       dispval.set(dispval.get() + text)
       inp.update()


root = Tk()
root.geometry("450x800")
root.title("CALCULATOR")


dispval = StringVar()
dispval.set("")
inp = Entry(root,textvariable=dispval,font="ALGERIAN 45 ")
inp.pack(fill=X,padx=20,pady=25,ipadx=9)

f1 = Frame(root,bg="grey")
b = Button(f1, text="9",padx=23,pady=18,font="30",borderwidth=10,relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1, text="8",padx=23,pady=18,font="30",borderwidth=10,relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1, text="7",padx=23,pady=18,font="30",borderwidth=10,relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)
f1.pack()

f1 = Frame(root,bg="grey")
b = Button(f1,text="6",padx=23, pady=18, font="30", borderwidth=10,relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1,text="5",padx=23,pady=18,font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1, text="4",padx=23,pady=18,font="30", borderwidth=10,relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)
f1.pack()

f1 = Frame(root,bg="grey")
b =Button(f1,text="3",padx=23,pady=18,font="30",borderwidth=10,relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1,text="2",padx=23,pady=18,font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1, text="1",padx=23,pady=18,font="30", borderwidth=10,relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)
f1.pack()

f1 = Frame(root,bg="grey")
b = Button(f1, text="0", padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1, text="+", padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>', click)

b = Button(f1, text="-", padx=25, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>', click)
f1.pack()

f1 = Frame(root,bg="grey")
b = Button(f1, text="*", padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1, text="/", padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>', click)

b = Button(f1, text="=", padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>', click)
f1.pack()

f1 = Frame(root,bg="grey")
b = Button(f1, text="C",bg="yellow",padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>',click)

b = Button(f1, text="(", padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>', click)

b = Button(f1, text=")", padx=23, pady=18, font="30", borderwidth=10, relief=RAISED)
b.pack(side=LEFT,padx=9,pady=8)
b.bind('<Button-1>', click)
f1.pack()

root.mainloop()
```
